```python
# Loda librarries and data file .csv
import pandas as pd 
df = pd.read_csv("/Users/pkrosniak/Developer/drip_leaderboard/pairs.csv")

#create another df with wallet adresses 
df1 = df.loc[df["0"].str.startswith('0x', na=False)]
#remove rows with NaN/walet addreses from main df
df=df.dropna(subset=['1'])
# reasing index in both df and copy wallet adreses to main df 
df.index = df1.index
df['team_address']= df1['0']
df = df.drop(['0','1','2','3','4','5'], axis=1)
df.head()
```

```python
#format values in token columns 
df['total_deposits'] = df['total_deposits'].astype(float)/1000000000000000000
df['total_payouts'] = df['total_payouts'].astype(float)/1000000000000000000
df['airdrops_total'] = df['airdrops_total'].astype(float)/1000000000000000000
df['airdrops_received'] = df['airdrops_received'].astype(float)/1000000000000000000
```

```python
#round columns in token columns to 2 digits after . 
cols = ['total_deposits', 'total_payouts', 'airdrops_total', 'airdrops_received']
df[cols] = df[cols].round(2)
```

```python
#Team ranking will be calculated based on Number of Airdrop Received - Airdrop Send / Number of Directs 
#In example below, it will be 34.197 - 32.382 / 51 = 0.035
df['referrals'] = df['referrals'].astype(str).astype(float)
df['ranking'] = df.airdrops_total - df.airdrops_received / df.referrals
df['ranking'] = df['ranking'].round(2)
```
